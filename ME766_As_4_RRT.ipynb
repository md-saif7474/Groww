```python
import matplotlib.pyplot as plt
import numpy as np
import math


def checkCollision(a, b, c, x = 14, y = 14, radius = 2):
      
    # Finding the distance of line 
    # from center.
    dist = ((abs(a * x + b * y + c)) /
            math.sqrt(a * a + b * b))
  
    # Checking if the distance is less 
    # than, greater than or equal to radius.
    if (radius == dist):
        return True
    elif (radius > dist):
        return True
    else:
        return False

def dist(a,b):
  x1 = a[0]
  y1 = a[1]
  x2 = b[0]
  y2 = b[1]
  return ((x1-x2)**2 + (y1-y2)**2)**.5


(x_s,y_s) = (1,1)
(x_d,y_d) = (27,27)
circle = plt.Circle((14,14), radius = 2, color = 'green')
rectangle = plt.Circle((20,20), radius = 3 , facecolor = "orange", edgecolor ="black")
fig, ax = plt.subplots(figsize=(10,10))
plt.plot(x_s,y_s, 'C0o')
plt.plot(x_d,y_d, 'C1o')

ax.add_patch(circle)
ax.add_patch(rectangle)
d = 4
theta = math.atan((y_d-y_s)/(x_d-x_s))
p = [[x_s,y_s, theta]]
q = [x_d, y_d]
flag = True
k = 0
while(flag):
  for point in p:
    if(dist(point, q)<=d):
      plt.plot(x_d,y_d,'o',color='black')
      plt.plot([point[0],x_d],[point[1],y_d], color = 'brown')
      flag = False
      print("Reached")
      break
    else:
      temp = point
      p.pop(0)
      x = point[0]
      y = point[1]
      theta = point[2]
      theta1 = theta -40*math.pi/180
      theta2 = theta +40*math.pi/180
      for th in [theta1,theta, theta2]:
        new_x = x + d*math.cos(th)
        new_y = y + d*math.sin(th)
        # print(new_x, new_y)
        while ((new_x-14)**2+(new_y-14)**2)<=4 or ((new_x-20)**2+(new_y-20)**2)<=9 or checkCollision((new_x-x),(y-new_y),y*(x-new_x)+x*(new_y-y)) or checkCollision((new_x-x),(y-new_y),y*(x-new_x)+x*(new_y-y),20,20,3):
          th = th +40*math.pi/180
          # print("Hello")
          new_x = x + d*math.cos(th)
          new_y = y + d*math.sin(th)
          # print(new_x, new_y)
        plt.plot([x,new_x],[y,new_y], color='brown')
        plt.plot(new_x,new_y,'o', color = 'red')
        # print("hello")
        th = math.atan((new_y-y_d)/(new_x-x_d))
        p.append([new_x, new_y, th])
#ax.add_patch(point)

#plt.axis('off')
plt.show() 
```
